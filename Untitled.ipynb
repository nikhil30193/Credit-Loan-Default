```python
cont_col = ['disbursed_amount','asset_cost','ltv','Date.of.Birth','PERFORM_CNS.SCORE','PRI.CURRENT.BALANCE','PRI.SANCTIONED.AMOUNT',
            'PRI.DISBURSED.AMOUNT','SEC.CURRENT.BALANCE','SEC.SANCTIONED.AMOUNT','SEC.DISBURSED.AMOUNT','SEC.INSTAL.AMT',
            'AVERAGE.ACCT.AGE','CREDIT.HISTORY.LENGTH']
```

```python
X = data.drop('loan_default',axis=1)
y = data['loan_default']
```

```python
from scipy import stats
pd.Series([stats.f_oneway(X[i],y) for i in cont_col], index = cont_col)
```

```python
cat_col = X.columns
cat_col.drop(cont_col)
cat_col
```

```python
import scipy.stats as stats
from scipy.stats import chi2_contingency

Significance = pd.Series()

for i in cat_col:
    dfObserved = pd.crosstab(X[i],y)
    chi2, p, dof, expected = stats.chi2_contingency(dfObserved.values)
    Significance[i] = (1-p)*100

Significance.sort_values()
```

```python
from sklearn.preprocessing import scale
X = scale(X)
```

```python
from sklearn.model_selection import train_test_split
train_X,test_X,train_y,test_y=train_test_split(X,y,test_size=.2,random_state=1)
```

```python
from sklearn.linear_model import LogisticRegression
lg = LogisticRegression()
```

```python
lg.fit(train_X,train_y)
lg.score(train_X,train_y)*100
```
